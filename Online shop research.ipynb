# Исследование интернет-магазина компьютерных игр "Стримчик"

**"Стримчик"** - онлайн-магазин компьютерных игр, продающий их по всему миру. 
Необходимо найти *потенциально популярный продукт* и спланировать *рекламные кампании*. Для этого нужно проанализировать некоторые закономерности, определяющие успешность игры.
Имеются данные за **2016** год, используя их, надо определить стратегический план на **2017** год.

Имеется файл с данными `games_csv` со следующей информацией:
<br>
`Name` — название игры
<br>
`Platform` — платформа
<br>
`Year_of_Release` — год выпуска
<br>
`Genre` — жанр игры
<br>
`NA_sales` — продажи в Северной Америке (миллионы проданных копий)
<br>
`EU_sales` — продажи в Европе (миллионы проданных копий)
<br>
`JP_sales` — продажи в Японии (миллионы проданных копий)
<br>
`Other_sales` — продажи в других странах (миллионы проданных копий)
<br>
`Critic_Score` — оценка критиков (максимум 100)
<br>
`User_Score` — оценка пользователей (максимум 10)
<br>
`Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

<b>План проекта:</b>
1. Изучим данные
2. Подготовим информацию к анализу
3. Проведем исследовательский анализ
4. Составим портрет пользователя каждого региона
5. Проверим некоторые гипотезы
6. Сделаем выводы

### Динамика изменения продаж в зависимости от платформы

Выберем платформы с наиболее высокими суммарными продажами и рассмотрим изменение количества продаж в зависимости от года.

```python
# создадим сводную таблицу с количесвом продаж по платформе
highest_sales = (data.pivot_table(index='platform', values='total_sales', aggfunc='sum')
                 .sort_values(by='total_sales', ascending=False)).reset_index()
highest_sales.head(15)
```

Заметим, что наибольшие продажи с большим отрывом от других всего у 6 платформ - `PS2`, `X360`, `PS3`, `Wii`, `DS`, `PS`. Рассмотрим их.

Составим сводную таблицу, которая содержит минимальный и максимальный годы выпуска игр по каждой платформе, встречающиеся в датафрейме.

## Проверка гипотез

### Исследование пользовательских рейтингов платформ Xbox One и PC

**Выдвинем следующие гипотезы:**
- H0: Средний пользовательский рейтинг платформы *Xbox One* **равен** среднему пользовательскому рейтингу платформы *PC*
- H1: Средний пользовательский рейтинг платформы *Xbox One* **не равен** среднему пользовательскому рейтингу платформы *PC*

Рассмотрим распределение оценок пользователя внутри платформ:

```python
xone = new_data.query('platform == "XOne" & user_score > -1')
pc = new_data.query('platform == "PC" & user_score > -1')
print('Платформа XOne:\n\n', xone['user_score'].describe())
print('\n\nПлатформа PC:\n\n', pc['user_score'].describe())
```

У выборок разные дисперсии и размеры, поэтому в т-тесте установим параметр `equal_var=False`.  Уровень статистической значимости установим 5%. Проведем т-тест с использованием метода `st.ttest_ind()`, который используется для независимых выборок, как и в нашем случае.

```python
result = st.ttest_ind(xone['user_score'], pc['user_score'], equal_var=False)
alpha = 0.05
print('pvalue:', result.pvalue)

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Нулевую гипотезу о равенстве средних оценок пользователя платформ не удалось отвергнуть. Значит, очень большая вероятность, что средние пользовательские рейтинги платформ `XOne` и `PC` действительно равны.
